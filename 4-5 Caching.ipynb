```python
from langchain.chat_models import ChatOllama
from langchain.prompts import PromptTemplate, ChatPromptTemplate, load_prompt
from langchain.callbacks import StreamingStdOutCallbackHandler
from langchain.globals import set_llm_cache, set_debug
from langchain.cache import InMemoryCache, SQLiteCache #cashing at memory and sqlite
```

```python
# set_llm_cache(InMemoryCache())
set_llm_cache(SQLiteCache("./cache/cache.db"))
set_debug(True)
```

```python
ollama = ChatOllama(
    model = "mistral:latest",
    temperature=0.1,
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()]
)
```

```python
ollama.predict("How can I fill a fountain pen with ink") #no caching - 1m 34.2s
```

```python
ollama.predict("How can I fill a fountain pen with ink") # caching - 0.0s
```
